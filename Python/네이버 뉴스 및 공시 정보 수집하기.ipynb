# 전체 과정을 하나의 함수로 만들기

```python
import FinanceDataReader as fdr
import numpy as np
import time
```

```python
df_krx = pd.read_csv("../00. 강의안/krx.csv")
df_krx.head()
```

```python
def get_news(item_name, start_page, end_page):
    news_list = []
    
    # KRX 자료를 통해 종목 코드 찾기
    item_code = df_krx.loc[df_krx["Name"] == item_name, "Symbol"].tolist()[0]
    
    # start_page 부터 end_page까지 뉴스/공시 자료 수집
    for page_no in range(start_page, end_page + 1):
        url = f"https://finance.naver.com/item/news_news.nhn?code={item_code}&page={page_no}"
        news = pd.read_html(url, encoding="cp949")
        temp = news[0]
        # 불필요한 데이터 제거
        df_news = temp[temp["제목"].str.contains("연관") == False]
        df_news = df_news.reset_index(drop=True)
        news_list.append(df_news)
        random_sleep = np.random.uniform(0, 1)
        time.sleep(random_sleep)
    df = pd.concat(news_list, ignore_index=True)
    return df
```

```python
# get_news 함수를 통해 '하이브'의 뉴스/공시 자료 수집

get_news("하이브", 1, 3)
```
