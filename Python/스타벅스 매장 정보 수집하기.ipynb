# 전체 과정을 하나의 함수로 만들기

```python
def get_store_list(sido_name, gugun_name=""):
    sido = sido_list.loc[sido_list["sido_nm"] == sido_name, "sido_cd"].tolist()[0]
    if len(gugun_name) > 0 :
        gugun = gugun_list.loc[gugun_list["gugun_nm"] == gugun_name, "gugun_cd"].tolist()[0]
    else:
        gugun = gugun_name
    url = "https://www.starbucks.co.kr/store/getStore.do"
    params = f"in_biz_cds=0&in_scodes=0&ins_lat=37.56682&ins_lng=126.97865&search_text=&p_sido_cd={sido}&p_gugun_cd={gugun}&isError=true&in_distance=0&in_biz_cd=&iend=1000&searchType=C&set_date=&rndCod=2U1ZHFAIQ8&all_store=0&T03=0&T01=0&T27=0&T12=0&T09=0&T30=0&T05=0&T22=0&T21=0&T10=0&T36=0&T43=0&T48=0&P10=0&P50=0&P20=0&P60=0&P30=0&P70=0&P40=0&P80=0&whcroad_yn=0&P90=0&new_bool=0"
    response = requests.post(f"{url}?{params}")
    SB = pd.DataFrame(response.json()['list'])[cols]
    return SB
```
